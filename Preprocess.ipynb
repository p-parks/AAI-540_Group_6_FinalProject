```python
df = pd.read_csv("TREC_07.csv")
df.head()
```

```python
# Drop the receiver email address as well as the date fields. 
df2 = df.drop(["receiver", "date"], axis=1)
df2.head()
```

```python
df2.info()
```

```python
# Check for rows where there are null or NaN fields.
null_data = df2[df2.isna().any(axis=1)]
null_data
```

```python
# Check for duplicates
duplicates = df2[df2.duplicated(subset=None, keep=False)]
print(duplicates)
```

```python
# Convert \n to spaces in the body text
df2['body'] = df2['body'].str.replace(r'\n', ' ', regex=True)
df2.head()
```

```python
# Save df2 to a CSV file.
df2.to_csv('TREC_07_cleaned.csv', index=False)
```

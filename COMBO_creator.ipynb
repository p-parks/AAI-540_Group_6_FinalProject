```python
import pandas as pd
```

```python
ceas = pd.read_csv("CEAS_08_cleaned.csv")
print(ceas.shape)
ceas.head()
```

```python
trec07 = pd.read_csv("TREC_07_cleaned.csv")
print(trec07.shape)
trec07.head()
```

```python
# Separate the HAM from TREC_07
trec_ham = trec07[trec07['label'] == 0]
print(trec_ham.shape)
trec_ham.head()
```

```python
combo = pd.concat([ceas, trec_ham], ignore_index=True)
print(combo.shape)
combo.head()
```

```python
combo.to_csv('combo.csv', index=False)
```

```python
import pandas as pd
```

```python
df = pd.read_csv("CEAS_08_cleaned.csv")
df
```

```python
# Create a new DataFrame for HAM emails
ham_df = df[df['label'] == 0]

# Create a new DataFrame for SPAM emails
spam_df = df[df['label'] == 1]

# Save the HAM DataFrame as a new CSV file
ham_df.to_csv('ham.csv', index=False)

# Save the SPAM DataFrame as a new CSV file
spam_df.to_csv('spam.csv', index=False)
```
